```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# 创建地图
plt.figure(figsize=(8, 8))
m = Basemap(projection='ortho', resolution=None, lat_0=35, lon_0=103)
m.bluemarble(scale=0.5)

# 在地图上画出兰州市的经纬度
lon, lat = 103.8, 36.06
x, y = m(lon, lat)
plt.plot(x, y, 'ok', markersize=5)
plt.text(x, y, 'Lanzhou', fontsize=12)

# 显示地图
plt.show()
```
